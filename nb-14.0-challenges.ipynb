### Challenge 1: 
Execute the code cell below to see an example of how it works. 
Use markdown in the cell after the code-block to describe the function `random_words_api`. Try to be descriptive about what each step of code in this function does, and why it works. 

```python
def random_words_api(nwords=10):
    "no docstring"
    URL = "https://random-word-api.herokuapp.com/word"
    response = requests.get(url=URL, params={"number": nwords})
    return response.json()

# demonstration
random_words_api(5)
```

**Description:** 

The fuction prints out a designated amout of random words from the specified url 


### Challenge 2: 
Use the `random_words_api` function to get 50 random words and store the result as a variable. Write a function that takes the list of words as input and returns a dictionary with the longest word as the key and the length of the longest word as a value. If there is a tie in the length of words then have it return additional words as keys with their lengths as values.

```python
api50_list = random_words_api(50)
#print(api50_list)

def longest_word(word_list): 
    """
    Create a dictionary and return the longest word and its length
    :param word_list: the list of words to be passed to the function
    """
    # Find the length of the longest string in the list
    longest_str_len = len(max(word_list, key=len))    
    
    # Loop through each word in word list and create a list of those matching the length of the longest word.  
    len_match_word_list = [word for word in word_list if len(word) == longest_str_len]
    
    # Loop through each word in word list and create a list of lengths for those matching the longest word.
    len_match_len_list = [len(word) for word in word_list if len(word) == longest_str_len]
    
    # Create a dict from the two lists using the words as the keys and lengths as the values.
    final_dict = dict(zip(len_match_word_list, len_match_len_list))
    
    # Return the final dict
    return(final_dict)
    
longest_word(api50_list)

```

### Challenge 3: 
Write a function to take the list of words as input and trim all words to be at most 5 characters in length, and return as a list.

```python
api50_list = random_words_api(50)

#print(api50_list)

def trim_to_five(word_list):
    """
    Take a list of strings, trim them to len == 5 and return
    :param word_list: the list of words to be passed to the function
    """
    # create an empty list
    empty_list = []
    # loop through each word in the passed word_list
    for word in word_list:
        # trim the word to include a maximum of five characters
        trimmed_word = word[0:5]
        # add the trimmed_word to the empty list
        empty_list.append(trimmed_word)
    
    # return the empty_list
    return(empty_list)
        
trim_to_five(api50_list)
```

### Challenge 4: 
Write a function to take a list of words as input and to count the occurrence of all letters in every word and return as a dictionary mapping letters to integers., e.g., {'a': 10, 'b': 3, 'c': 5, ...}.  

```python
api50_list = random_words_api(50)

def character_count(word_list):
        """
        counts the the characters in a list
        """
        # Import thelowercase alphabet from the string library
        from string import ascii_lowercase
        alphabet = ascii_lowercase
        
        # Create one long string from the list of words input
        string_for_analysis = ''.join(word_list)
        
        # Create an empty dict to be populated by the letter and count
        alphabet_counts = {}
        
        # Loop through each letter in the alphabet
        for letter in alphabet:
            # Count the occurrence of the letter in the string
            char_count = string_for_analysis.count(letter)
            # Create a dict with the letter and number of occurrences as the key and value
            char_count_dict = {letter: char_count}
            # Add this dict to the empty dict
            alphabet_counts.update(char_count_dict)
            
        return(alphabet_counts)
        
character_count(api50_list)
```

### Challenge 7:

```python
carry_on = True

iterations = 0

while carry_on == True:
    iterations = iterations + 1
    random_number = np.random.randint(low = 1, high = 100)
    #print(random_number)
    
    if random_number < 25 and random_number > 22:  
        break
        
print(f'Complete in {iterations} iterations')
```
